```python
from sklearn.preprocessing import LabelEncoder

# Create a copy
df_encoded = df.copy()

# Encoding Lifestyle_Habits
le_lifestyle = LabelEncoder()
df_encoded['Lifestyle_Encoded'] = le_lifestyle.fit_transform(df['Lifestyle_Habits'].astype(str))
print("Lifestyle Mapping:", dict(zip(le_lifestyle.classes_, le_lifestyle.transform(le_lifestyle.classes_))))
df = df_encoded
```

```python
df.info()
```

```python
required_columns = [
    "Age", "Sleep_Hours", "Study_Hours", "Screen_Time_Post_10PM", 
    "Physical_Activity_Minutes", "Mood_Score", "Sleep_Debt",
    "Stress_Level", "Stress_Per_Study_Hour", "Burnout_Risk_Score",
    "Lifestyle_Encoded" , "Overall_Wellbeing_Score","Sleep_Quality_Score",
    "Mental_Fatigue_Score"
]

df = df[required_columns]
```

```python
df
```

```python
X = df[['Age',  'Sleep_Hours', 'Study_Hours',
        'Screen_Time_Post_10PM', 'Physical_Activity_Minutes', 'Mood_Score',
        'Sleep_Debt', 'Lifestyle_Encoded']]

# Selecting target/output variable
Y = df[['Burnout_Risk_Score','Stress_Level', 'Stress_Per_Study_Hour',"Overall_Wellbeing_Score","Sleep_Quality_Score",
    "Mental_Fatigue_Score"]]
```

```python
X , Y
```

```python
#train test split

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.35, random_state=42
)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score

from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor

# Multi-output model setup
multi_model = MultiOutputRegressor(RandomForestRegressor(random_state=42))
multi_model.fit(X_train, Y_train)

# Predictions
Y_pred = multi_model.predict(X_test)

   
r2 = r2_score(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)

print(f"    RÂ² Score:           {r2:.4f}")
print(f"    Mean Absolute Error: {mae:.4f}")
print(f"    Mean Squared Error:  {mse:.4f}")
```
